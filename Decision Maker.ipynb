```python
# Replace latitude and longitude values from those in the console
# Default Berkeley coordinates
latitude = 37.866454399999995
longitude = -122.2525292
```

```python
search = input("What are you looking for? ")
```

```python
headers = {'Authorization': 'Bearer %s' % api_key}
r = requests.request('GET', f"https://api.yelp.com/v3/businesses/search?term={search}&latitude={latitude}&longitude={longitude}", headers=headers, params=None)
```

```python
r.url
```

```python
places = r.json()  
pprint(places)
```

```python
# we only want places that are open, at least 4/5 stars, and close enough to walk or take the bus (within a mile/1609 meters)
new_places = []
for business in places['businesses']:
  if business['rating'] >= 4 and not business['is_closed'] and business['distance'] < 1609:
    new_places.append(business)
```

```python
def tts(text, filename="tts.mp3"):
  gTTS(text).save(filename)
  return IPython.display.display(IPython.display.Audio(filename, autoplay=True))
```
